```python
import pandas as pd
import numpy as np
import plotly.plotly as py
import plotly.graph_objs as go 
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
```

```python
df=pd.read_csv('history.csv')
```

```python
data = dict(type = 'choropleth',
            locations = df['Country Code'],
            colorscale='Green',
            text= df['CPI 2016 Score'],
            z=df['CPI 2016 Score'],
            colorbar = {'title':'CPI Index'})
```

```python
layout = dict(
    title = '2016 CPI',
    geo = dict(
        showframe = False,
        
    )
)
```

# World CPI Index

Corruption perception Index for the year 2016 the red color indicates it is not corrupted as the color goes down it indicates more corrupted

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap,validate=False)
```

# CPI Improvement in 2016 compared to 2015

The below data shows information map of countries who have showed a improvement in their CPI index from 2015

```python
df9=df[(df['CPI 2016 Score']>df['CPI 2015 Score'])]
```

```python
df9
```

```python
data = dict(type = 'choropleth',
            locations = df9['Country Code'],
            colorscale='Green',
            text= df9['CPI 2016 Score'],
            z=df9['CPI 2016 Score'],
            colorbar = {'title':'CPI Index'})
```

```python
layout = dict(
    title = '2016 CPI improving Countries',
    geo = dict(
        showframe = False,
        
    )
)
```

```python
choromap = go.Figure(data = [data],layout = layout)
iplot(choromap,validate=False)
```

# CPI factor with respect to region

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sns.set_style('ticks')

sns.violinplot(x='Region',y='CPI 2016 Score',data=df,alpha=0.9,estimator=np.mean,linewidth=3,)
sns.set_context('poster',font_scale=10)
```

When we see the graph above we can see that Sahara-African region is the most corrupted according to the data, Europe and Central Asia region has the highest mean which relates to the fact that the corruption is less when compared to other regions
